```python
import json
import os
import pandas as pd
import matplotlib.pyplot as plt

results = []

for file in os.listdir("../results"):
    if file.endswith(".json"):
        with open(f"../results/{file}") as f:
            results.append(json.load(f))
```

## Model comparison

```python
df = pd.DataFrame(results)
pd.set_option("display.max_colwidth", None)
display(df)
```

### Closer look to each column

```python
flattened_dfs = {}

for col in df.columns:
    
    # ignorer best_params
    if col == "best_params":
        continue

    # vérifier si la colonne contient des dictionnaires
    if df[col].apply(lambda x: isinstance(x, dict)).any():
        flat_df = pd.json_normalize(df[col])
        flat_df.index = df["model"]

        name = f"df_{col}"
        flattened_dfs[name] = flat_df
        globals()[name] = flat_df

        print(f"\n{name}")
        display(flat_df)
```
